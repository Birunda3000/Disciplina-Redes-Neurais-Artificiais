```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

from google.colab import drive
from IPython.display import display
```

```python
drive.mount('/content/drive')
diretorio = '/content/drive/MyDrive/Redes Neurais/DATA/'
#diretorio2 = '/content/drive/MyDrive/Redes Neurais/'
```

```python
data = pd.read_csv(diretorio+"subsfin.txt", sep=' ', header=None)#.to_numpy()
data
```

```python
Realizando normalização
```

```python
scaler = MinMaxScaler()
data = scaler.fit_transform(data)
df = pd.DataFrame(data)
df
```

Series a serem trabalhadas

```python
janela, graficos = plt.subplots(nrows=13, ncols=4, figsize=(15, 35))
plt.tight_layout(pad=1.0)
subplots = janela.axes

for i in df.columns:
    df_new = pd.DataFrame(df[i])
    subplots[i].plot(df_new)
    subplots[i].set_title(f'Sensor {i+1}');
```

Suavizando outliers

```python
df_clean = pd.DataFrame()
for serie in df.columns.values:
    df_clean[serie] = pd.DataFrame(df[serie])
    out = (df_clean[serie] - np.mean(df_clean[serie])) / np.std(df_clean[serie])
    idx = np.where((np.abs(out) >= 3) | (np.abs(out) <= -3))
    df_clean[serie].iloc[idx] = np.mean(df_clean[serie])
```

Exibindo series com outlires substitutidos pela media

```python
janela, graficos = plt.subplots(nrows=13, ncols=4, figsize=(15, 35))
plt.tight_layout(pad=1.0)
subplots = janela.axes

for i in df_clean.columns:
    df_new = pd.DataFrame(df_clean[i])
    subplots[i].plot(df_new)
    subplots[i].set_title(f'Sensor {i+1}');
```

Fazendo janelamento e separando os targets

```python
time_window = 100
X = []
y = []
for i in range(len(data)-time_window):
    X.append(data[i:i+time_window])
    y.append(data[i+time_window])
X = np.array(X)
y = np.array(y)
X.shape, y.shape
```

Separando teste e treino

```python
X_train, y_train = X[:8640], y[:8640]
X_test, y_test = X[-5760:], y[-5760:]
X_test.shape, y_test.shape, X_train.shape, y_train.shape
```
